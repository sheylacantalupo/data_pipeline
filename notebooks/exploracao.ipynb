```python
import json
```

```python
with open(path_json, 'r') as file:
    # carregar dados JSON de um arquivo 
    dados = json.load(file)
```

```python
print(type(dados))
```

```python
# como é inviável trabalhar os dados dessa forma, preciso ter acesso ter acesso ao conteúdo que está na chave "raw_lines"
for chave, valor in dados.items():
    print(f'Chave: {chave}, Valor: {valor}')
```

```python
# O tipo de raw_lines é uma lista que contém todo o conteúdo dos produtos(lista de strings)
raw_lines = dados['payload']['blob']['rawLines']
print(type(raw_lines))
print(type(raw_lines[100]))
print(len(raw_lines))
```

```python
# Será necessário realizar o tratamento de cada linha para assim, utilizar o método json.load(que aceita apenas o formato válio json)
raw_lines = raw_lines[1:-1] # Retirei os colchetes que estavam em excesso, eliminando o primeiro e último caractere
print(type(raw_lines))
# Observe que cada linha contém os dados de um produto, porém observe que os elemento dessa lista não está no formato padrao json
# existe uma vírgula no fim de cada string
print(raw_lines[0][:-1])
```

```python
dados_json = [json.loads(line[:-1]) for line in raw_lines]
print(type(dados_json))
print(type(dados_json[1]))

print(dados_json[1]['Nome do Produto'])
```
